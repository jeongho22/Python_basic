```python
# 데이터 불러오기
df = pd.read_csv("../pythonwork/ai/data/data1/train.csv")
```

```python
df =df.drop('Cabin', axis=1)
```

```python
df.shape
df.info()
```

```python
df['Sex'] = df['Sex'].map({'male':0, 'female':1})

df.head()
```

```python
df =df.dropna()

df

df.isnull().sum()

df.describe()
```

```python
y = df['Survived'].values

print(y)
```

```python
x = []

for i, rows in df.iterrows():
    x.append([rows['Fare'],rows['Age'],rows['Pclass'],rows['Sex']]) # 요금 나이 등급
    
print(x)
```

```python
###텐서 플로우 안에 케라스를쓰면 딥러닝이 쉬워진다... w1,w2이런거구하기쉬움
# activation 활성 함수는 내마음대로
# sigmoid는 0~1에서 확률을 뱉고싶을때 쓰는함수


#1. 모델 디자인
# 히든 레이어 1,2,3 개쓴거임 지금

model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(128, activation='tanh'),   #히든레이어임 이게...
    tf.keras.layers.Dense(256, activation='tanh'),  # 노드의 갯수 64,128,1개 동그란거갯수
    tf.keras.layers.Dense(1, activation='sigmoid'), #마지막에는 하나의 노드가 되어야함..
])


# 2. model complie 하기 # ?? 파라미터(매개변수) 넣기
# binary_crossentropy 0~1사이에서 확률문제에서 씀

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# model.fit(x데이터,y데이터, epochs=10)

model.fit(np.array(x),np.array(y), epochs=1000) # test값 x
```

```python
# 예측 
# 요금,나이,객실등급,성별: 1여자 0남자

pr = model.predict([[15,22,3,1],[15,29,3,0]])
print(pr)

#즉 티켓의 요금이 높고 어리고 높은등급객실에 여자가 살 확률이 존나게 높다.
```
